```python
#function for assigning k-mers & bag of words
class sequence_process:
  def __init__(self, kmersize,df):
    self.size = kmersize
    self.df = df
    self.vectored_df = df
    self.encoded_virus_class = []
    self.virus = []
  
  def GetKmers(self,sequence):
    return [sequence[x:x+self.size] for x in range(len(sequence) - self.size + 1)]
  
  def apply_kmers(self):
    kmer = []
    for i in self.df.sequence:
      kmer.append(self.GetKmers(i))
    kmer = np.array(kmer)
    self.df['kmers'] = kmer
    return self.df
  
  def long_string(self,sequence):
    join_letter = ' '
    return(join_letter.join(sequence))
  
  def apply_joining(self):
    all_joined_kmers = []
    for i in self.df.kmers:
      join_kmers = self.long_string(i)
      all_joined_kmers.append(join_kmers)
    all_joined_kmers = np.array(all_joined_kmers)
    self.df['kmers'] = all_joined_kmers
    self.df['kmers'] = self.df['kmers'].astype('str')
    return self.df

  def vectorize_features(self):
    from sklearn.feature_extraction.text import CountVectorizer
    from sklearn.preprocessing import LabelEncoder
    self.apply_kmers()
    self.apply_joining()

    cv = CountVectorizer() #ngram_range=(4,4)
    vectoring_kmer = list(self.df['kmers'])
    X = cv.fit(vectoring_kmer)
    Transformed = X.transform(vectoring_kmer)
    print('Total number of features detected :',len(X.get_feature_names_out()))
    total_feature = Transformed.toarray()
    new_df = pd.DataFrame(total_feature ,columns = X.get_feature_names_out() )
    self.vectored_df = new_df
    self.vectored_df['genus_name'] = self.df['genus_name']
    LE = LabelEncoder()
    self.encoded_virus_class = LE.fit_transform(self.vectored_df ['genus_name'])
    self.virus = LE.inverse_transform(self.encoded_virus_class)
    self.vectored_df ['genus_name'] = self.encoded_virus_class
    return self.vectored_df

  def virus_name(self):
    return(self.encoded_virus_class , self.virus)

  def train_test(self):
    from sklearn.model_selection import train_test_split
    X_train, X_test, y_train, y_test = train_test_split(self.vectored_df.drop('genus_name', axis = 1), self.vectored_df['genus_name'], test_size = 0.2, random_state=42)
    X_train, X_val, y_train, y_val = train_test_split(X_train,  y_train, test_size = 0.2, random_state=42)
    return X_train, X_val, X_test, y_train, y_val, y_test
```
