```python
#import libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
final_df.drop('Unnamed: 0', axis = 1, inplace = True, )
```

```python
from sklearn.preprocessing import LabelEncoder
LE = LabelEncoder()
encoded_virus_class = LE.fit_transform(final_df['genus_name'])
#virus = LE.inverse_transform(encoded_virus_class)
final_df['genus_name_label'] = encoded_virus_class
```

Deep Learning algorithm - ANN

```python
from sklearn.preprocessing import OneHotEncoder
targets = final_df['genus_name']
encoder = OneHotEncoder()
reshaped_data = targets.values.reshape(-1, 1)
encoded_data = encoder.fit_transform(reshaped_data)
encoded_dataframe = pd.DataFrame(encoded_data.toarray().astype('int64').tolist())
```

```python
encoded_dataframe
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(final_df.drop(['genus_name','genus_name_label'], axis = 1), encoded_dataframe, test_size = 0.2, random_state=42)
X_train, X_val, y_train, y_val = train_test_split(X_train, y_train, test_size = 0.2, random_state=42)
```

```python
from sklearn.feature_selection import VarianceThreshold
selector = VarianceThreshold(threshold = 0)
X_train = selector.fit_transform(X_train)
X_val = selector.transform(X_val)
X_test = selector.transform(X_test)
```

```python
import numpy as np
import tensorflow as tf
from tensorflow import keras
from keras import layers
from sklearn.model_selection import KFold
```

```python
import numpy as np
import tensorflow as tf
from tensorflow import keras
from keras import layers
from sklearn.model_selection import KFold
from keras.layers import Conv2D, MaxPooling2D, Flatten, Dense, Dropout
from keras.models import Sequential

# Define the Keras Sequential model
model = Sequential()
model.add(layers.Dense(512, activation='relu', input_shape=(8618,)))


model.add(Dropout(0.5))
model.add(layers.Dense(256, activation='relu'))
#model.add(Dropout(0.5))
model.add(layers.Dense(128, activation='relu'))
model.add(layers.Dense(64, activation='relu'))

model.add(layers.Dense(27, activation='softmax'))
optimizer_1 = keras.optimizers.legacy.Adam(learning_rate=0.001)
model.compile(optimizer=optimizer_1, loss='categorical_crossentropy', metrics=['accuracy']) #categorical_crossentropy
model.summary()
```

```python
# Train the model
history = model.fit(X_train, y_train, epochs=300, batch_size=32, validation_data=(X_val, y_val))
```

```python
loss, accuracy = model.evaluate(X_test, y_test)
predict = model.predict(X_test)
# Print the accuracy
print("Test accuracy:", "{:.2f}".format(accuracy))
```

```python
# Load the data
x_train = X_train.values
y_train = y_train.values

# Define the number of folds
n_folds = 5

# Initialize the cross-validation folds
kf = KFold(n_splits=n_folds, shuffle=True, random_state=42)

# Iterate through the folds
fold = 0
for train_index, val_index in kf.split(x_train):
    fold += 1
    print(f'Fold {fold}')

    # Get the training and validation data for this fold
    x_train_fold, y_train_fold = x_train[train_index], y_train[train_index]
    x_val_fold, y_val_fold = x_train[val_index], y_train[val_index]

    # Train the model on this fold
    history = model.fit(x_train_fold, y_train_fold, epochs=30, validation_data=(x_val_fold, y_val_fold))
```

```python
loss, accuracy = model.evaluate(X_test, y_test)

# Print the accuracy
print("Test accuracy:", "{:.2f}".format(accuracy)) #
```
