```python
f = open("sequences.txt", "w")
for i in range(len(fasta_sequences)):
    if 'OS' in fasta_sequences[i].description:
        index_1 = fasta_sequences[i].description.index('OS=')
        seq = fasta_sequences[i].seq
        name = fasta_sequences[i].description[index_1+3:]
        index_2 = name.index(' ') #change to 'OX' for species level resolution
        name = name[:index_2]
        f.write('>' + name +'\n')
        f.writelines(seq + '\n')
f.close()
```
